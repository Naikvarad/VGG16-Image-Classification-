```python
from keras.applications.vgg16 import VGG16
```

```python
Model = VGG16()
```

```python
print(Model.summary())
```

```python
from keras.utils import load_img
```

```python
image = load_img(r"C:\Users\Varad S N\OneDrive\Desktop\MUG.jpeg",target_size=(224,224))
```

```python
from keras.utils import img_to_array
```

```python
image = img_to_array(image)
```

```python
image.shape
```

```python
image = image.reshape(1,image.shape[0],image.shape[1],image.shape[2])
```

```python
image.shape
```

```python
from keras.applications.vgg16 import preprocess_input
```

```python
image = preprocess_input(image)
```

```python
image
```

```python
Y_pred_Lable = Model.predict(image)
```

```python
Y_pred_Lable
```

```python
from keras.applications.vgg16 import decode_predictions
```

```python
lable = decode_predictions(Y_pred_Lable)
```

```python
lable
```

```python
Final=lable[0][0]
```

```python
print('%s(%.2f%%))' %(Final[1],Final[2]*100))
```
